<h4> Построим функцию распределения случайной величины на основе выборки.</h4>

<h3>Сгенерируем три выборки для случайных расстояний между двумя точками, равномерно распределенные в прямоугольнике. Получим среднее значение расстояния между точками, построить
функцию распределения вероятностей и плотности вероятностей случайных
расстояний.
</h3>

```python
def distance(x1, y1, x2, y2):
    return np.sqrt((x2-x1)**2+(y2-y1)**2)
```

```python
def distance_line(number):
    a=10
    b=30
    distances=[]
    for i in range(number):
        x1=np.random.uniform(0,a)
        y1=np.random.uniform(0,b)
        x2=np.random.uniform(0,a)
        y2=np.random.uniform(0,b)
        distances.append(distance(x1, y1, x2, y2))
    distances.sort()
    return distances
```

```python
graf100=distance_line(100)
graf1000=distance_line(1000)
graf10000=distance_line(10000)
print (np.average(graf100))
print (np.average(graf1000))
print (np.average(graf10000))
```

```python
def fun_uniform(x, a, b): 
    return ((x-a)/(b-a));

def fun_uniform_d( a, b): 
    return (1/(b-a));
```

```python
a=10
b=30

uniform_x=np.linspace(a, b, 100)
uniform_y=fun_uniform(uniform_x, a, b)


uniform_d_x= [a, b]
uniform_d_y=[fun_uniform_d( a, b), fun_uniform_d( a, b)]
```

```python
plt.figure(dpi=100)
plt.plot(uniform_x, uniform_y, label='uniform')
plt.plot(uniform_d_x, uniform_d_y, label='uniform_d')
plt.legend()
plt.show()
```
